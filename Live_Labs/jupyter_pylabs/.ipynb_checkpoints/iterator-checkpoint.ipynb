```python
text = "welcome"
it = iter(text)   # converting str object into an object of iter 
while True:
    try:
        print(next(it))
    except StopIteration:
        break
```

```python
class CountUp:
    def __init__(self, max):
        self.max = max
        self.current = 1

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.current > self.max:
            raise StopIteration
        val = self.current
        self.current += 1
        return val

for num in CountUp(5):
    print(num)
 
```

```python
class EvenNumbers:
    def __init__(self, limit):
        self.limit = limit
        self.num = 0
 
    def __iter__(self):
        return self
 
    def __next__(self):
        if self.num > self.limit:
            raise StopIteration
        value = self.num
        self.num += 2
        return value
 
for n in EvenNumbers(10):
    print(n)
```
