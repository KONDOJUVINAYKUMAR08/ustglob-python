```python
#4. File not found error
try:
    f = open("test.txt", "r")
    data = f.read()
except FileNotFoundError:
    print("File not found.")
finally:
    print("Execution completed.")
```

```python
#5. key Error
try:
    my_dict = {"a": 1}
    print(my_dict["b"])
except KeyError:
    print("Key does not exist.")
```

```python
#6. IndexError
try:
    lst=[1,2,4]
    print(lst[5])
except IndexError as e:
    print(e)
```

```python
#7. Custom Exception
class NegativeNumberError(Exception):
    pass
try:
    num = -5
    if num<0:
        raise NegativeNumberError("Negative numbers are not allowed.")
except NegativeNumberError as e:
    print(e)
```

```python
class NegativeNumberError(Exception):     # Exception is predefined class. It is base class 
    pass                                  # pass does nothing
try:
    age=int(input("Enter value for age: ")) 
    if age < 0:
        raise NegativeNumberError("Age with negative value is not allowed")
    else:
        print("Age is   " , age )
except NegativeNumberError as e:
    print("Custom Exception:", e)
```

```python
class NegativeNumberError(Exception):     # Exception is predefined class. It is base class 
    pass                                  # pass does nothing
try:
    age=int(input("Enter value for age: ")) 
    if age < 0:
        raise NegativeNumberError("Age with negative value is not allowed")
    if age >= 0 and age<18:
        raise NegativeNumberError("Age must be 18 or above 18.")
    if age>=18:
        print("My age is: ",age)
except NegativeNumberError as e:
    print("Custom Exception:", e)
```
