In Python, a generator is a special type of function that returns values one at a time, instead of returning all values at once.
 
It uses the yield keyword.
 
Simple Definition
 
A generator is a function that produces values one by one using yield.
 
def count_up(n):

    for i in range(1, n + 1):

        yield i
 
for num in count_up(5):

    print(num)
 
 
How It Works
 
yield returns a value and pauses the function
 
Next call resumes from where it stopped
 
Values are generated lazily (only when needed)
 
Generator vs Normal Function
 
Normal Function		Generator
 
Uses return		Uses yield
 
Returns once		Returns many times
 
Stores all values	Generates values on demand
 
 

```python
import sys
squares_list = [x*x for x in range(1_000_000)]
squares_gen = (x*x for x in range(1_000_000))
print("List size:", sys.getsizeof(squares_list))     
print("Generator size:", sys.getsizeof(squares_gen))
 
 
```
