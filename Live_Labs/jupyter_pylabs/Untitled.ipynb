```python
data= [(1,2),(3,1),(0,4)]
sorted_by_first = sorted(data)  # Default sorts by the first element
print(sorted_by_first)
sorted_by_second = sorted(data, key=lambda x: x[1])
print(sorted_by_second)
```

```python
data= [(1,2),(3,1),(0,4)]
sorted_by_first = sorted(data, reverse=True)  # Default sorts by the first element
print(sorted_by_first)
sorted_by_second = sorted(data, key=lambda x: x[1], reverse= True)
print(sorted_by_second)
```

```python
data2 = [(1, 3), (1, 2), (2, 1)]
sorted_by_first_second = sorted(data2, key=lambda x: (x[0], x[1]))
print(sorted_by_first_second)
```

```python
data = { 'date': 1, 'apple': 5, 'banana': 2, 'cherry': 7}
sort_by_key=dict(sorted(data.items()))
print(sort_by_key)
```

```python
data = {'apple': 5, 'banana': 2, 'cherry': 7, 'date': 1}
sort_by_key=dict(sorted(data.items() , key=lambda x: x[0]))
print(sort_by_key)
sort_by_value=dict(sorted(data.items() , key=lambda x: x[1]))
print(sort_by_value)
```

```python
data = {'apple': 5, 'banana': 2, 'cherry': 7, 'date': 1}
sort_by_key=dict(sorted(data.items() , key=lambda x: x[0], reverse=True))
print(sort_by_key)
sort_by_value=dict(sorted(data.items() , key=lambda x: x[1], reverse=True))
print(sort_by_value)
```
