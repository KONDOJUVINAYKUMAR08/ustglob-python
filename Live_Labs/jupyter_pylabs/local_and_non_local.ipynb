```python
def outer():
    x = 10
    def inner():
        x = 20   # This creates a NEW local variable
        print(x)
    inner()
    print(x)

outer()   # 10,1,2,7,3,4,5,8
#output : 20 ,10
# Python treats a variable assigned inside a function as local to that function,
# even if a variable with the same name exists outside it.
 
```

```python
def outer():
    x = 10
    def inner():
        x=x+1     # we can access variable declared outside but can not be
        print(x)  # modified  
 
    inner()
    print(x)
 
outer()   # 10,1,2,7,3,4,5,8
```

```python
def outer():
    x = 10
    def inner():
        nonlocal x 
        x=x+1     # we can access variable declared outside but can not be
        print(x)  # modified  
 
    inner()
    print(x)
 
outer()   # 10,1,2,7,3,4,5,8
```

```python
def counter():
    count=0
    def increment():
        nonlocal count
        count+=1
        return count
    return increment

c=counter()
print(c)
print(c())
print(c())
```

Key Rules
 
nonlocal works only in nested functions
 
It cannot refer to global variables
 
It refers to the nearest enclosing functionâ€™s variable
 
 
Simple Definition
 
A nonlocal variable is a variable defined in an outer function that can be modified inside an inner function 
using the nonlocal keyword.
